## Data visualization

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))
plt.title('Churns')

plot_distribution(consumption, 'cons_12m', axs)
```

```python
correlation_matrix2 = price_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix2, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Remarks:
- Higher median of non-churned suggests that longer contract durations may correlate with customer retention.
- Shorter contract durations might be associated with higher churn rates (0.13).

```python
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'])
client_df['date_end'] = pd.to_datetime(client_df['date_end'])

client_df['contract_duration'] = (client_df['date_end'] - client_df['date_activ']).dt.days

plt.figure(figsize=(10, 7))
sns.boxplot(x='churn', y='contract_duration', data=client_df)

plt.title('Churn vs Contract Duration')
plt.xlabel('Churn')
plt.ylabel('Contract Duration (Days)')
plt.show()


client_df['contract_duration_bins'] = pd.cut(client_df['contract_duration'], bins=5)
churn_rate_by_duration = client_df.groupby('contract_duration_bins')['churn'].mean()

plt.figure(figsize=(10, 6))
churn_rate_by_duration.plot(kind='bar', color='coral')
plt.title('Churn Rate by Contract Duration')
plt.xlabel('Contract Duration (Binned)')
plt.ylabel('Churn Rate')
plt.show()
```

Remarks:
- Off peak price of energy is the highest.
- Off peak price of energy is about 4 times higher than mid peak price of energy.

```python
sns.lineplot(data = price_df, x='price_date', y='price_off_peak_var', label = 'off peak', color = 'blue')
sns.lineplot(data = price_df, x='price_date', y='price_peak_var', label = 'peak', color = 'green')
sns.lineplot(data = price_df, x='price_date', y='price_mid_peak_var', label = 'mid peak', color = 'purple')

plt.title('Energy Price Distribution by season')
plt.xlabel('Period')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
